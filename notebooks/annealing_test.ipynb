```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros((75,3))
```

```python
A[0, :] = [1,0,0]
```

```python
sh_step = 0.1
h_step = 0.01
h_max = 0.5
```

```python
for i in range(1, 75):
    A[i, 1:] = A[i-1, 1:] + [sh_step, h_step]
    if A[i, 2] > h_max:
        A[i, 2] = h_max
    A[i, 0] = 1 - np.sum(A[i, 1:])
    v = A[i, :]
    v[v > 1] = 1
    v[v < 0] = 0
    if v.sum() > 1:
        v[v.argmax()] = v[v.argmax()] - (v.sum() - 1)
    A[i, :] = v
```

```python
plt.plot(A)
```

```python
def LinearAnnealingProbabilities(actual, sh_step=0.1, h_step=0.01, h_max=0.5):
    v = np.zeros(3)
    v[1:] = actual[1:] + [sh_step, h_step]
    if v[2] > h_max:
        v[2] = h_max
    v[0] = 1 - np.sum(v[1:])
    v[v > 1] = 1
    v[v < 0] = 0
    if v.sum() > 1:
        v[v.argmax()] = v[v.argmax()] - (v.sum() - 1)
    return v
```

```python
A = np.zeros((75, 3))
A[0, :] = [1, 0, 0]
for i in range(1, 75):
    A[i, :] = LinearAnnealingProbabilities(A[i-1, :], sh_step=0.1)
```
